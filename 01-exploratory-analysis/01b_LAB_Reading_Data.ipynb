```python
# Create a path to the database
path = 'data/baseball.db'

# Create a connection to the database
con = sq3.Connection(path)

# Create a query to select all of the rows from the allstarfull table
query = 'SELECT * FROM allstarfull;'

# Use the connection to run the query and return the results to a DataFrame
observations = pds.read_sql(query, con)
display(observations.head())
```

```python
# Compute the number of games using the batting table and an average starting position using the allstarfull table
query = '''
SELECT 
    b.playerID,
    SUM(b.G) AS num_games_played,
    AVG(a.startingPos) AS avg_starting_position,
    SUM(b.G) / 10000 * 0.5 + (10 - AVG(a.startingPos)) * 0.5 AS metric
FROM batting b
JOIN allstarfull a ON b.playerID = a.playerID
WHERE a.startingPos IS NOT NULL
GROUP BY b.playerID
ORDER BY metric DESC
LIMIT 10;
'''

display(pds.read_sql(query, con))
```
